## Write a query to get total fine paid per Office

```python
col = [' cash_fine ', ' online_fine ']
fine_df[col] = fine_df[col].replace({'\$': '', ',': ''}, regex=True)
fine_df[col] = fine_df[col].replace({' -   ':"0"," NA ":0})

#fine_df[col] = fine_df[col].replace(' -   ', '').astype(int)

fine_df[' cash_fine '] = pd.to_numeric(fine_df[' cash_fine '])
fine_df[' online_fine '] = pd.to_numeric(fine_df[' online_fine '])
fine_df.fillna(0)
```

```python
fine_df.head()
```

```python
fine_df['Total_fine'] = fine_df[' cash_fine '] + fine_df[' online_fine ']
```

```python
fine_df.groupby(by="org_indiv").agg({"Total_fine":"sum"})
```

## Write a query to get the highest total fine and lowest total fine paid per state

```python
fine_per_state = fine_df.groupby(by="state").agg({"Total_fine":"sum"})
```

```python
fine_per_state.max()
```

```python
fine_per_state.min()
```

```python
fine_df.groupby(by="state").agg({" online_fine ":"sum"}).sort_values(by=" online_fine ")[:2]
```

## Write a SQL query to find the number of doctors that contributed to the state having highest 'cash' fine

```python
fine_df.groupby(by="state").agg({" cash_fine ":"sum", "first_name":"count"}).sort_values(by=" cash_fine ", ascending=False)[:1]['first_name']
```

## Get the top 5 average fine paid per state and category

```python
fine_df.groupby(by="state").agg({" cash_fine ":"mean"}).sort_values(by=" cash_fine ", ascending=False)[:5]
```

```python
fine_df.groupby(by="category").agg({" cash_fine ":"mean"}).sort_values(by=" cash_fine ", ascending=False)[:5]
```
